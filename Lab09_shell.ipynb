```python
## Name: Nicholas Almeida
## Student Number: 200385
def evenlySpacedValues(interval,n):
    a = interval[0]
    b = interval[1]
    L = [a] # will hold the list of points
    for i in range(n-1):
        L.append(L[i]+(b-a)/(n-1))
    return L
def LagrangePoly(i,points):
    L = 1
    for j in range(len(points)):
        if(i != j):
            L *= (x-points[j][0])/(points[i][0]-points[j][0])
    return expand(L)
## Write your ClosedNewtonCotes function here
def ClosedNewtonCotes(f, a, b, n):
    #function body
    result = 0
    xvalues = evenlySpacedValues([a,b],n+1)
    yvalues = [f(x=value) for value in xvalues]
    if n == 1:
        result = (b-a)*(yvalues[0]+yvalues[1])/2
        return result.n()
    elif n == 2:
        result = (b-a)/6*(f(x=a)+4*f(x=xvalues[1])+f(x=b))
        return result.n()
    elif n == 3:
        result = (b-a)/8*(f(x=a)+3*f(x=xvalues[1])+3*f(x=xvalues[2])+f(x=b))
        return result.n()
    elif n == 4:
        result = (b-a)/90*(7*f(x=a)+32*f(x=xvalues[1])+12*f(x=xvalues[2])+32*f(x=xvalues[3])+7*f(x=b))
        return result.n()
    else:
        # return the general Closed Newton-Cotes formula
        # first compute the coefficients
        coeff =[]
        try:
            for i in range(n+1):
                coeff.append(f(x=xvalues[i]))
        except:
            print("Error: f(x) is not defined at the point x = ", xvalues[i])
            return
        # next find the lagrange polynomials
        try:
            # put x and y values into a list of tuples
            points = []
            for i in range(n+1):
                points.append((xvalues[i],yvalues[i]))
            # find the lagrange polynomials
            lagrange = []
            for i in range(n+1):
                lagrange.append(LagrangePoly(i,points))
        except:
            print("Error: LagrangePoly function failed at i = ", i)
            return
        # finally compute the integral
        try:
            for i in range(n+1):
                result += coeff[i]*lagrange[i]
        except:
            print(" Failed to multiply coeff " , coeff[i], " and lagrange ", lagrange[i].n())
        # integrate the result on the interval [a,b]
        result = integrate(result,(x,a,b))
        result = result.n()
        return result
        


        
```
